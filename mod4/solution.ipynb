# Data Transformations

```python

# Deal with missing values
# remove columns that add no value
# Scale and fix skew problem 
# One-hot encode categorical
# Remove potential outliers
# Feature engineering


df = data.copy()
cat_df = df.select_dtypes(include='object')
num_df = df.select_dtypes(include='number')
```

```python
# Drop rows which have multiple columns missing
df.dropna(inplace=True, thresh=21)
print(f'Original data length: {len(data)}. Now {len(df)}')
```

```python
# Fix Date 
df['Sale Date'] = pd.to_datetime(df['Sale Date'],format='%Y-%M-%d')
```

```python
# Drop redundant columns
df.drop(columns=['Owner Name','State','Suite/ Condo   #','image','Address','Legal Reference','Property City'],inplace=True,errors='ignore')
```

```python
def remove_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3-Q1 
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    outlier_mask = ~((col<lower_bound)|(col>upper_bound))
    return col[outlier_mask]

print('before:',num_df.isnull().sum())
num_df = num_df.apply(remove_outliers)
print('after:',num_df.isnull().sum())
num_df.dropna(how='any', inplace=True)
```

## Feature Engineering

```python
#### Extract address type
my_dict = {'DR':'Drive',
           'AVE':'Avenue',
           'RD':'Road',
           'ST':'Street'}

def extract_address_type(address):
    if not pd.isna(address):
        split_address = address.split(' ')
        for word in split_address:
            if word in my_dict:
                return my_dict.get(word)
            
        return 'Not Available'
cat_df['Property_Address_Extract'] = cat_df['Property Address'].apply(extract_address_type)
cat_df.drop(columns=['Property Address'], axis=1, inplace=True)
```

```python
print(len(cat_df))
for col in cat_df:
    value_counts = cat_df[col].value_counts()
    remove_low_counts = value_counts[value_counts<5].index
    cat_df = cat_df[~cat_df[col].isin(remove_low_counts)]
print(len(cat_df))
```

```python
df['Sale Month'] = df['Sale Date'].dt.month
df['Sale Year'] = df['Sale Date'].dt.year
df['Sale Day'] = df['Sale Date'].dt.day
df.drop('Sale Date', axis=1, inplace=True)
```

```python
# One-hot encoding
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
ohe_transformed = ohe.fit_transform(cat_df)

# Combining dfs
comb_df = pd.concat([ohe_transformed, num_df],axis=1)
```

```python
comb_df.dropna(how='any',inplace=True)
```

```python
comb_df.drop('Total Value',axis=1, inplace=True)
```

```python
def evaluate_model(model, features, y_test, y_pred, name):
    fig, axs = plt.subplots(ncols=1, nrows=2, figsize=(10,20))
    y_pred_flat = y_pred.flatten()
    y_test_flat = y_test.flatten()

    sns.scatterplot(x=y_test_flat, y=y_pred_flat, ax=axs[0])
    axs[0].set_title(f'{name} Model Actual Vs. Predicted Values')
    axs[0].set_ylim(0,600000)
    axs[0].set_xlim(0,600000)
    plt.savefig(f'{name} ActualVsPredicted.png')
    try:
        coefs = model.coef_.flatten()
    except AttributeError:
        coefs = model.feature_importances_
        
    
    model_feature_imp = pd.DataFrame({'Coefficients':coefs}, index=features).sort_values(by='Coefficients',ascending=True)
    model_feature_imp.plot(kind='barh', ax=axs[1])
    axs[1].set_title(f'{name} Model Feature Importance')

    plt.tight_layout()
    plt.savefig(f'{name} Model Importance.png')

    mse_ = mse(y_test_flat, y_pred_flat)
    rmse = np.sqrt(mse_)
    print(f'{name} Model RMSE:', rmse)
    
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
X = comb_df.drop('Sale Price',axis=1)
y = comb_df['Sale Price'].values.reshape(-1,1)

pt_features = PowerTransformer(method='yeo-johnson',standardize=True)
pt_target =   PowerTransformer(method='yeo-johnson', standardize=True)
X_transformed = pt_features.fit_transform(X)
y_transformed = pt_target.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_transformed, y_transformed, test_size=0.2, random_state=0)
print(f'{len(X_train)} training data points')
print(f'{len(X_test)} testing data points')
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse


linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

linear_y_pred_transformed = linear_model.predict(X_test)
linear_y_pred = pt_target.inverse_transform(linear_y_pred_transformed)
y_test_original = pt_target.inverse_transform(y_test)

evaluate_model(linear_model, X.columns,y_test_original , linear_y_pred, 'Linear')
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=1000, random_state=0)
y_train_flat = y_train.ravel()
rf.fit(X_train, y_train_flat)
rf_y_pred_transformed = rf.predict(X_test).reshape(-1,1)
rf_y_pred = pt_target.inverse_transform(rf_y_pred_transformed)
evaluate_model(rf, X.columns, y_test_original, rf_y_pred, 'Random Forest')
```

# Gradient Boosting Model

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
y_train_flat = y_train.ravel()
gb_model.fit(X_train, y_train_flat)

gb_y_pred_transformed = gb_model.predict(X_test).reshape(-1,1)
gb_y_pred = pt_target.inverse_transform(gb_y_pred_transformed)
evaluate_model(gb_model, X.columns, y_test_original, gb_y_pred, 'Gradient Boosting Model')
```
